```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import StandardScaler
from sklearn.gaussian_process.kernels import *
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
df_X_train = pd.read_csv('X_train.csv')
df_y_train =  pd.read_csv('y_train.csv')
df_X_test =  pd.read_csv('X_test.csv')
```

```python
df_X_train = df_X_train.drop(columns=['id'])
y_train = df_y_train.drop(columns=['id'])
df_X_test = df_X_test.drop(columns=['id'])
```

```python
df_X_test.shape
```

```python
df_X_test.head()
```

```python
df_X_train.head()
```

```python
imputer = KNNImputer(n_neighbors=2)

X_train_i = imputer.fit_transform(df_X_train)
X_test_i = imputer.fit_transform(df_X_test)


X_train_im = pd.DataFrame(X_train_i, columns=df_X_train.columns)
X_test_im = pd.DataFrame(X_test_i, columns=df_X_test.columns)
```

```python
lof = LocalOutlierFactor(n_neighbors=2)
pred_train = lof.fit_predict(X_train_im)
pred_test = lof.fit_predict(X_test_im)
```

```python
X_train_im['outlier_lof'] = pred_train
X_test_im['outlier_lof'] = pred_test
y_train['outlier_lof'] = pred_train
```

```python
X_train_im
```

```python
X_train_im = X_train_im[X_train_im.outlier_lof != -1]
X_test_im = X_test_im[X_test_im.outlier_lof != -1]
y_train = y_train[y_train.outlier_lof != -1] 
```

```python
X_train_f = X_train_im.drop(columns=['outlier_lof'])
y_train = y_train.drop(columns=['outlier_lof'])
X_test_f = X_test_im.drop(columns=['outlier_lof'])
```

```python
X_test_f.shape #thats a problem
```

```python
np.seterr(invalid='ignore')
new_datax = X_train_f
#feature selection train
selector = SelectKBest(f_regression, k=200)
new_datax_reduced = selector.fit_transform(new_datax, np.array(y_train).ravel())
cols = selector.get_support(indices=True)
```

```python
#scale 
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(new_datax_reduced)
X_scaled = scaler.transform(new_datax_reduced)
```

```python
kernel = 1.0 * RBF(1.0)
kernel = RationalQuadratic()
regressor = GaussianProcessRegressor(kernel=kernel)
```

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(regressor, X_train_f, y_train, cv=5,
                    scoring=('r2'),
                    return_train_score=True,
                    n_jobs=4)
```
