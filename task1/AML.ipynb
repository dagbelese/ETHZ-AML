```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import StandardScaler
from sklearn.gaussian_process.kernels import RBF
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
df_X_train = pd.read_csv('/Users/sanjanawarambhey/Downloads/task1/X_train.csv')
df_y_train =  pd.read_csv('/Users/sanjanawarambhey/Downloads/task1/y_train.csv')
df_X_test =  pd.read_csv('/Users/sanjanawarambhey/Downloads/task1/X_test.csv')
df_sample =  pd.read_csv('/Users/sanjanawarambhey/Downloads/task1/sample.csv')
```

```python
df_X_train = df_X_train.drop(columns=['id'])
y_train = df_y_train.drop(columns=['id'])
df_X_test = df_X_test.drop(columns=['id'])
```

```python
imputer = KNNImputer(n_neighbors=2)

X_train_i = imputer.fit_transform(df_X_train)
X_test_i = imputer.fit_transform(df_X_test)


X_train_im = pd.DataFrame(X_train_i, columns=df_X_train.columns)
X_test_im = pd.DataFrame(X_test_i, columns=df_X_test.columns)
```

```python
lof = LocalOutlierFactor(n_neighbors=2)
pred_train = lof.fit_predict(X_train_im)
pred_test = lof.fit_predict(X_test_im) 

X_train_im['outlier_lof'] = pred_train
X_test_im['outlier_lof'] = pred_test
y_train['outlier_lof'] = pred_train

X_train_im = X_train_im[X_train_im.outlier_lof != -1]
X_test_im = X_test_im[X_test_im.outlier_lof != -1]
y_train = y_train[y_train.outlier_lof != -1] 


X_train_f = X_train_im.drop(columns=['outlier_lof'])
y_train = y_train.drop(columns=['outlier_lof'])
X_test_f = X_test_im.drop(columns=['outlier_lof'])
```

```python
selector = SelectKBest(score_func=f_regression, k=200)
X_train_s = selector.fit_transform(X_train_f, np.array(y_train).ravel())

X_test_s = selector.transform(X_test_f)
X_train_se = pd.DataFrame(X_train_s)
X_test_se = pd.DataFrame(X_test_s)
```

```python
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train_se)
X_test_sc = scaler.transform(X_test_se)


X_train = pd.DataFrame(X_train_sc, columns=X_train_se.columns)
X_test = pd.DataFrame(X_test_sc, columns=X_test_se.columns)
```
